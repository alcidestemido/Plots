# Exemplos de como fazer gráfico no python  

```python
# Usando o comando figure e subplot do matplotlib

#Primeiro iremos definir nossas variaveis:
t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2*np.pi*t)
s2 = np.cos(4*np.pi*t)
#comando que define o tamanha do gráfico plt.rcParams['figure.figsize'] = (largura,comprimento) 
plt.rcParams['figure.figsize'] = (20,10)
```

```python
#Agora iremos plotar 4 graficos 
plt.figure() # comando de entrada 
plt.subplot(2,2,1) #subplot(numero de linhas , numero de colunas , posição do plot) 
plt.plot(t, s1) 
plt.subplot(2,2,2)
plt.plot(t, 2*s2)
plt.subplot(2,2,3)
plt.plot(t,4*s2)
plt.subplot(2,2,4)
plt.plot(t,10*s1)
```

#  A partir de uma função definida 

```python
def f(x):
    return x**2 - 30
```

```python
x1 = np.arange(-5, 5, 0.01)
ax = plt.subplot(1,1,1)
plt.plot(x1,f(x1))
plt.xticks(np.arange(min(x1), max(x1)+1, 1)) # quantidades de numeros no eixo x  ( valor inicial , final , subdivisões)
plt.yticks(np.arange(-50, 50, 4)) #quantidades de numeros no eixo y ( valor inicial , final , subdivisões)
plt.margins(0.05)  # zoom no grafico 
ax.spines['left'].set_position('center') # deslocando o eixo y para o centro da imagem 
ax.spines['bottom'].set_position('center')# deslocando o eixo x para o centro da imagem 
```

Exemplo do Matplotlib de um plot dentro do outro 

```python
# Plot inside a plot
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))
x = np.linspace(-0.50, 1., 1000)

# Outer Plot
ax.plot(x, x**3)
ax.plot(x, np.sin(x))
ax.set(xlim=(-0.5, 1.0), ylim=(-0.5,1.2))
fig.tight_layout()

# Inner Plot
inner_ax = fig.add_axes([0.2, 0.55, 0.35, 0.35]) # x, y, width, height
inner_ax.plot(x, x**2)
inner_ax.plot(x, np.sin(x))
inner_ax.set(title='Zoom In', xlim=(-.2, .2), ylim=(-.01, .02), 
             yticks = [-0.01, 0, 0.01, 0.02], xticks=[-0.1,0,.1])
ax.set_title("Plot inside a Plot", fontsize=20)

```
